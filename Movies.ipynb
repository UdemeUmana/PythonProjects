```python
import pandas as pd
import numpy as np
```

```python
#Importing my dataset
#the dropna() is used to drop any value that is not a number or not a time,etc
movies= pd.read_csv(r'c:\Users\Udeme Umana\Downloads\movies - movies (1).csv').dropna()
```

```python
#Displaying the first five columns in your dataset
movies
```

```python
#Displaying the number of rows and columns in my dataset
movies.shape
```

```python
#Printing information about my dataset
movies.info()
```

```python
#converting the the gross,budget,votes and runtime datatype to an integer
movies['gross']=movies['gross'].astype('int64')
movies['budget']=movies['budget'].astype('int64')
movies['votes']=movies['votes'].astype('int64')
movies['runtime']= movies['runtime'].astype('int64')
```

```python
#Getting the year from the released field and storing it in a new column(year)
movies[['Date','year']]= movies['released'].str.split(',' ,expand = True)
movies
```

```python
#describing basic statistics of the dataframe
movies.describe()
```

```python
#Displaying the last five rows of my dataset
movies.tail(50)
```

```python
#Finding the highest gross pay
movies['gross'].max()
```

```python
#Finding the lowest gross pay
movies['gross'].min()
```

```python
#Getting the unique values from the rating columns in a list format
list(movies['rating'].unique())
```

```python
#Finding the country with the highest vote in movies
movies1= movies.groupby('country')['votes'].count().reset_index()
movies1.sort_values(by='votes',ascending = False)
```

```python
#Finding the genre that has the hihest vote
movies1= movies.groupby('genre')['votes'].sum().reset_index()
movies1.sort_values(by='votes',ascending = False)
```

```python
#Displaying movies that the country is the United State and the genre is comedy
movies[(movies['country']=='United States') & (movies['genre']=='Comedy')]
```

```python
#Displaying the name of the movies and their budget in desc order to find the movie with the highest budget
name_budget= movies.groupby('name')['budget'].sum().reset_index()
name_budget.sort_values(by='budget',ascending = False)
```

```python
#sorting my dataset by the gross amount to find the country and movies name with the lowest gross amount
movies.sort_values('gross',ascending=True)
```
